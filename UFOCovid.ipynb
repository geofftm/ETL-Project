```python
import pandas as pd
import re
from sqlalchemy import create_engine
```

```python
ufo_file = "scrubbed.csv"
ufo_df = pd.read_csv(ufo_file)
ufo_df.columns = ufo_df.columns.str.replace(' ','_')
ufo_df.columns = ufo_df.columns.str.replace('(','')
ufo_df.columns = ufo_df.columns.str.replace(')','')
ufo_df.head()
```

```python
covid_file = "COVID19_state.csv"
covid_df = pd.read_csv(covid_file)
covid_df.columns = covid_df.columns.str.replace(' ','_')

#covid_df = covid_df.drop(columns=['Gini', 'ICU Beds', 'GDP', 'Unemployment', 'Sex Ratio', 'Smoking Data', 'Flu Deaths', 'Respiratory Deaths', 'Physicians', 'Hospitals', 'Health Spending', 'Pollution', 'Med-Large Airports', 'Temperature', 'Urban', 'Age 0-25', 'Age 26-54'])
covid_df.head()
```

```python
state_file = "csvData.csv"
state_df = pd.read_csv(state_file)
state_df.head()
```

```python
#state_df['lower'] = state_df['Code'].str.lower()
state_df = state_df.rename(columns={'State': 'State_Name'})

state_df.head()
```

```python
ufo_df = ufo_df.dropna(subset=['state'])
ufo_df.head()
```

```python
ufo_df['latitude'] = ufo_df.latitude.str.replace(r"[a-zA-Z]",'')
ufo_df.head()
```

```python
ufo_df['latitude'] = ufo_df['latitude'].astype('float')
ufo_df.head()
```

```python
ufo_df.dtypes
```

```python
engine = create_engine('postgresql://dummydummy')
```

```python
engine.table_names()
```

```python
ufo_df.to_sql(name='ufo', con=engine, if_exists='replace', index=False)


#new_customer_data_df.to_sql(name='customer_name', con=engine, if_exists='append', index=False)
```

```python
covid_df.to_sql(name='covid19', con=engine, if_exists='replace', index=False)
```

```python
state_df.to_sql(name='state', con=engine, if_exists='replace', index=False)
```
